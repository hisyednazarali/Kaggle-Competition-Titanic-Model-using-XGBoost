```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.get_dummies(df.Sex)
```

```python
df1
```

```python
df2 = pd.concat([df,df1],axis = 'columns')
```

```python
df2.head()
```

```python
df2.drop(["Sex"],axis=1,inplace = True)
```

```python
df2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2.Embarked = le.fit_transform(df2.Embarked)
```

```python
df2.head()
```

```python
X  = df2.drop("Survived",axis =1)
```

```python
X
```

```python
y = df2["Survived"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from xgboost import XGBClassifier
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
xgb_model = XGBClassifier()
xgb_model.fit(X_train,y_train)
```

```python
y_pred_train = xgb_model.predict(X_train)
y_pred_test = xgb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_test,y_pred_test)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(xgb_model,X_test,y_test)
plt.show()
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgb_model = XGBClassifier()
param_grid = {"n_estimators":[1,5,10,15,20,40,50,100],'max_depth':[3,4,5,6],'gamma':[0,1,2,3]}
```

```python
grid = GridSearchCV(xgb_model,param_grid,cv=5,scoring="accuracy")
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
predictions = grid.predict(X_test)
```

```python
predictions
```

```python
print(classification_report(y_test,predictions))
```

```python
grid.best_estimator_.feature_importances_
```
